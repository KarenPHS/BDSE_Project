```python
from selenium import webdriver
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from time import sleep
import time
import numpy as np
import pandas as pd
from datetime import datetime
import csv

pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',4000)

options =webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--incognito")               
options.add_argument("--disable-popup-blocking ")
driver = webdriver.Chrome(options = options)


# my_dict = {"股票代號":[], "公司名稱":[], "終止上櫃日期":[], "備註":[]} 
my_dict = {"下櫃名單":[]}

# ----------------------------------------------
# 上櫃買中心
def visit():
    driver.get('https://www.tpex.org.tw/web/regular_emerging/deListed/de-listed_companies.php?l=zh-tw')
    

# 查找全部，輸入股票代號內容清除、年分 選擇全部、選擇下櫃原因 選擇全部    
def search():
    # 股票代號內容清除
    select_column = driver.find_element(By.CSS_SELECTOR, "form#listed_companies")
    select_code = select_column.find_elements(By.CSS_SELECTOR, "input#stk_code")
    select_code.clear()
    
#     print(select_column.text)

    # 年分 選擇全部
    select_year = select_column.find_elements(By.CSS_SELECTOR, "div.v-pnl.pt5")[0]
    select_year1 = select_year.find_elements(By.CSS_SELECTOR, "select")[0]
    select_year1.click()
    select_year2 = select_year1.find_elements(By.CSS_SELECTOR, "option")[0]
    select_year2.click()
    
#     print(select_year2.text)
    
    # 選擇下櫃原因 選擇全部  
    select_reason = select_column.find_elements(By.CSS_SELECTOR, "div.v-pnl.pt5")[1]
    select_reason1 = select_reason.find_elements(By.CSS_SELECTOR, "select")[0]
    select_reason1.click()
    select_reason2 = select_reason1.find_elements(By.CSS_SELECTOR, "option")[0]
    select_reason2.click()
    
#     print(select_reason2.text)
    
    # 按下查詢紐
    Buttom = select_column.find_elements(By.CSS_SELECTOR, "div.v-pnl")[0]
    Buttom1 = Buttom.find_elements(By.CSS_SELECTOR, "input")[1]
    Buttom1.click()

# def column():
#     next_page = driver.find_elements(By.CSS_SELECTOR, "div.v-pnl.pt5")[2]
# #     print(next_page.text)
#     # 找到表格    
#     select_column = next_page.find_elements(By.CSS_SELECTOR, "tbody")[0]    
#     print(select_column.text)
#     my_dict["全部"].append(select_column.text)
    
#     # 存成DataFrame                   
#     df = pd.DataFrame(data=my_dict)
#     df.to_csv('全部.csv', index=False)
    
def page():    
    next_page = driver.find_elements(By.CSS_SELECTOR, "div.v-pnl.pt5")[2]
#     print(next_page.text)
    # 每一頁(找到按下一頁的紐)
    next_page1 = next_page.find_elements(By.CSS_SELECTOR, "tbody")[1]
    next_page2 = next_page1.find_element(By.CSS_SELECTOR, "span")
    next_page3 = next_page2.find_elements(By.CSS_SELECTOR, "a")[-2]
    next_page3.click()       


  
    
def test():  
    p = 0
    while p < 56:
        next_page = driver.find_elements(By.CSS_SELECTOR, "div.v-pnl.pt5")[2]
        select_column = next_page.find_elements(By.CSS_SELECTOR, "tbody")[0]
#         return select_column

        table = select_column.text.replace('\n', " ").split(" ")
#         print(table)
        
        #把這頁表格內容list，存成4組list
#         result = list(np.array_split(table, 4))
        
#         i = 0
#         step = 4        
#         result1 = [table[i:i+step] for i in (0,len(table),step)]
#         i = i + step
#         print(result1)
        
        # 把每個欄位一個個讀出來
        j = 0
        for content in table:
            
            # 每一頁的股票代號、公司名稱、終止上櫃日期、備註，都不要!(因為第一排是0,1,2,3)               
            if j//4 != 0:               
               
                my_dict["下櫃名單"].append(content)
#                 print(content)                    


            else:
                pass
            
            j = j + 1
            
                       
        # 存成DataFrame                   
        df = pd.DataFrame(data=my_dict)
     
        df.to_csv('全部.csv', index=False, )
                
        page()
        p = p + 1         
   

        
def end():
    #關閉瀏覽器
    driver.quit()    
   
 

    
if __name__ == '__main__':
    visit()
#     init()
    search()
    test()
#     a = test()
    end()    
```
