```python
import requests, re, pprint, os, time
from bs4 import BeautifulSoup
from urllib import request
import pandas as pd
import csv

# 自訂標頭

My_headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36'}


# ---------------------------------------------
# 讀取股票代號檔案



path = "C:/Users/User/Desktop/code_list.csv"

code_list=''
with open(path,newline='') as f:
    rows=csv.reader(f)
    for i in rows:
        code_list=i



# 直接到電子資料查詢作業
# 用迴圈方式替換股票代碼
# 用迴圈方式替換年分
# 找出重編字樣
# 存成DataFrame
my_dict = {"name":[], "date":[], "財報是否重編":[]} 
    
def get_re_fin(stock_number,year):  
    url=f"https://doc.twse.com.tw/server-java/t57sb01?step=1&colorchg=1&co_id={stock_number}&year={year}&seamon=&mtype=A&"
    
    try:
        
        page = requests.get(url, headers = My_headers)
        time.sleep(5)
        
    # 用soup 利用CSS Selector 抓取內文    
        if "重編" in page.text:
            soup = BeautifulSoup(page.text,"lxml")
            time.sleep(3)
#             print(soup)
            content = soup.select("form table")
            content = content[1]
#             print(content)
        
            re_fin = 1
            my_dict["name"].append(stock_number)
            my_dict["date"].append(year)
            my_dict["財報是否重編"].append(re_fin)
            print(my_dict)
            
        elif "查無所需資料" in page.text:
            print("查無所需資料")
            pass
        
        else:
            print("沒有財報重編")
            pass
    
    
    except requests.exceptions.ConnectionError:
        print('ConnectionError -- please wait 3 seconds')
        
        time.sleep(3)
    except requests.exceptions.ChunkedEncodingError:
        print('ChunkedEncodingError -- please wait 3 seconds')
        
        time.sleep(3)    
    except:
        print('Unfortunitely -- An Unknow Error Happened, Please wait 3 seconds')
        
        time.sleep(3)
        
        

df = pd.DataFrame(data=my_dict)
df.to_csv("財報重編名單",index=False)
    
            
if __name__ == '__main__':
    for stock_number in code_list:
        for year in range(100,111):    
            get_re_fin(stock_number,year) 

    
        
```
